````python
import ast
import pandas as pd
from tqdm import trange
import torch

def generate_batched_response(model_name, df, batch_size=4, maxTokens=2048):
    model, tokenizer = load_qwen(model_name)
    responses = []
    token_counts = []

    for start in trange(0, len(df), batch_size):
        end = min(start + batch_size, len(df))
        batch = df.iloc[start:end]

        # prepare rendered texts
        rendered_prompts = []
        for _, row in batch.iterrows():
            question = row["question"]
            choices = row["options"]
            context = row["context"]

            if isinstance(choices, str):
                choices_list = "\n".join([f"{chr(65+i)}. {c}" for i, c in enumerate(ast.literal_eval(choices))])
            else:
                choices_list = "\n".join([f"{chr(65+i)}. {c}" for i, c in enumerate(choices)])

            messages = [
                {"role": "user", "content": f"""
                ### HƯỚNG DẪN
                Bạn là một chuyên gia về bộ môn Lịch Sử. Bạn sẽ được cung cấp một câu hỏi lịch sử và bốn lựa chọn A, B, C và D. **SỬ DỤNG NỘI DUNG ĐƯỢC CUNG CẤP**, trả lời bằng cách đưa ra đáp án trước, xong rồi giải thích đáp án của bạn sau.
                Tuyệt đối **KHÔNG** nhắc lại câu hỏi và các lựa chọn được cho. Quy tắc này tuân thủ NGHIÊM NGẶT.

                ---

                ### VÍ DỤ MẪU

                Nội dung: "Tháng 3 – 1947, Chính phủ Pháp cử Bôlae làm Cao uỷ Pháp ở Đông Dương, thực hiện kế hoạch tiến công căn cứ địa Việt Bắc nhằm nhanh chóng kết thúc chiến tranh. Pháp huy động 12 000 quân và hầu hết máy bay ở Đông Dương mở cuộc tiến công Việt Bắc từ ngày 7 – 10 – 1947. Sáng 7 – 10 – 1947, quân dù Pháp chiếm thị xã Bắc Kạn, thị trấn Chợ Mới. Cùng ngày, binh đoàn bộ binh từ Lạng Sơn theo Đường số 4 đánh lên Cao Bằng, rồi vòng xuống Bắc Kạn theo Đường số 3, bao vây Việt Bắc ở phía đông và phía bắc. Ngày 9 – 10 – 1947, một binh đoàn hỗn hợp từ Hà Nội ngược sông Hồng và sông Lô lên Tuyên Quang, Chiêm Hoá, bao vây Việt Bắc ở phía tây. Khi địch tiến công, Đảng ta có chỉ thị “Phải phá tan cuộc tiến công mùa đông của giặc Pháp”. Trên khắp các mặt trận, quân dân ta anh dũng chiến đấu. Quân ta chủ động bao vây, tiến công địch ở Chợ Mới, Chợ Đồn, Chợ Rã,... buộc Pháp phải rút khỏi Chợ Đồn, Chợ Rã cuối tháng 11 – 1947. Ở mặt trận hướng đông, quân ta phục kích chặn đánh địch trên Đường số 4, tiêu biểu là trận đèo Bông Lau (30 – 10 – 1947), thu nhiều vũ khí. Ở mặt trận hướng tây, quân dân ta phục kích đánh địch nhiều trận trên sông Lô, nổi bật là trận Đoan Hùng, Khe Lau, đánh chìm nhiều tàu chiến, ca nô của địch. Sau hơn hai tháng, đại bộ phận quân Pháp phải rút khỏi Việt Bắc ngày 19 – 12 – 1947. Quân dân ta loại khỏi vòng chiến đấu hơn 6 000 địch, bắn rơi 16 máy bay, bắn chìm 11 tàu chiến, ca nô. Cơ quan đầu não kháng chiến được bảo toàn. Bộ đội chủ lực của ta trưởng thành. Cuộc kháng chiến chuyển sang giai đoạn mới. Sau thất bại này, Pháp buộc phải thay đổi chiến lược từ “đánh nhanh, thắng nhanh” sang “đánh lâu dài”, thực hiện chính sách “dùng người Việt đánh người Việt, lấy chiến tranh nuôi chiến tranh”. Phối hợp với Việt Bắc, quân dân trên các chiến trường toàn quốc hoạt động mạnh, kiềm chế địch."

                Câu hỏi: Chiến thắng Việt Bắc thu-đông năm 1947 của quân dân Việt Nam đã

                Các lựa chọn: ['A. làm thất bại chiến tranh cục bộ.', 'B. làm thất bại chiến tranh tổng lực.', 'C. làm thất bại chiến tranh đặc biệt.', 'D. bảo vệ an toàn cơ quan đầu não kháng chiến.'

                Trả lời:
                ---

                ###
                Sau đây là câu hỏi bạn phải trả lời:

                Nội dung: {context}

                Câu hỏi: {question}

                Các lựa chọn: {choices_list}

                Trả lời:"""}
            ]
            # render into a single string prompt
            rendered_prompts.append(
                tokenizer.apply_chat_template(messages, tokenize=False)
            )

        # tokenize the batch of rendered prompts
        inputs = tokenizer(
            rendered_prompts,
            return_tensors="pt",
            padding=True,
            truncation=True
        ).to(model.device)

        # generate for the whole batch
        with torch.no_grad():
            outputs = model.generate(**inputs, max_new_tokens=maxTokens)

        # decode each sample
        for i in range(outputs.size(0)):
            input_len = (inputs["input_ids"][i] != tokenizer.pad_token_id).sum().item()
            response = tokenizer.decode(outputs[i][input_len:], skip_special_tokens=True)
            responses.append(response)
            token_counts.append(len(tokenizer(response)["input_ids"]))

    df["response"] = responses
    df["response_token_count"] = token_counts
    return df
````

````python
# Cell 3: Định nghĩa lớp LLMJudge (Phiên bản so sánh Benchmark hàng loạt - 8 tiêu chí)

class LLMJudge:
    def __init__(self, rubric_path: str, api_key: str, judge_model: str):
        if not api_key:
            raise ValueError("Không tìm thấy GOOGLE_API_KEY. Vui lòng thiết lập trong Cell 2.")

        genai.configure(api_key=api_key)
        self.model = genai.GenerativeModel(judge_model)
        self.metrics = self._load_rubric(rubric_path)
        self.metric_names = [m['name'] for m in self.metrics]

        print(f"✅ LLM Judge đã được khởi tạo với model: {judge_model}")
        print(f"✅ Đã tải thành công {len(self.metrics)} tiêu chí từ: {rubric_path}")
        print("⚙️ Chế độ: So sánh Benchmark hàng loạt (1 API call / 8 tiêu chí).")

    def _load_rubric(self, path: str) -> list:
        """Tải rubric từ file JSON."""
        with open(path, 'r', encoding='utf-8') as f:
            return json.load(f)

    # --- ĐÃ CẬP NHẬT PROMPT CHO 8 TIÊU CHÍ ---
    def _create_bulk_comparison_prompt(self, data_row: dict) -> str:
        """Tạo prompt để so sánh và đánh giá tất cả các tiêu chí cùng lúc."""

        criteria_text = "\n".join([f'- **{metric["name"]}**: {metric["description"]}' for metric in self.metrics])

        question = data_row.get('question', 'Không có')
        correct_answer = data_row.get('ans', 'Không có')
        benchmark_answer = data_row.get('benchmark', 'Không có')
        model_response = data_row.get('reponse', 'Không có')

        return f"""
Bạn là một chuyên gia đánh giá AI công tâm. Nhiệm-vụ của bạn là so-sánh "Câu trả-lời cần đánh-giá" với "Câu trả-lời mẫu (Benchmark)" và cho điểm cho TẤT CẢ các tiêu-chí được cung-cấp.

--- CÂU HỎI ---
{question}

--- ĐÁP ÁN ĐÚNG (tham khảo) ---
{correct_answer}

--- CÂU TRẢ LỜI MẪU (BENCHMARK) ---
{benchmark_answer}

--- CÂU TRẢ LỜI CẦN ĐÁNH GIÁ ---
{model_response}

--- BỘ TIÊU CHÍ ĐÁNH GIÁ ---
Hãy cho điểm "Câu trả lời cần đánh giá" so với "Câu trả lời mẫu (Benchmark)" theo thang điểm 1-5 cho mỗi tiêu chí sau:
{criteria_text}

--- NHIỆM VỤ CỦA BẠN ---
Câu trả lời của bạn BẮT BUỘC phải là một đối tượng JSON duy nhất, không có văn bản nào khác.
Đối tượng JSON phải chứa các key là tên của 8 tiêu chí, và value là điểm số (số nguyên từ 1 đến 5).

Ví dụ định dạng output:
{{
  "tính súc tích": 4,
  "tính dễ hiểu": 5,
  "tính chặt chẽ": 3,
  "tính liên quan": 5,
  "tính đúng đắn": 4,
  "tính đầy đủ": 3,
  "Cấu trúc câu trả lời": 4,
  "Định dạng câu trả lời": 5
}}
"""

    def _judge_all_criteria(self, data_row: dict) -> dict:
        """Gửi một yêu cầu duy nhất để đánh giá tất cả các tiêu chí."""
        prompt = self._create_bulk_comparison_prompt(data_row)
        default_scores = {name: 0 for name in self.metric_names}
        max_retries = 5
        attempt = 0

        while attempt < max_retries:
            try:
                response = self.model.generate_content(prompt)
                cleaned_text = response.text.strip().replace("```json", "").replace("```", "")
                judgement = json.loads(cleaned_text)

                if all(key in judgement for key in self.metric_names):
                    return judgement
                else:
                    attempt += 1
                    print(f"\n🟡 Lỗi: JSON trả về thiếu key. Thử lại... (lần {attempt}/{max_retries})")
                    time.sleep(1)

            except Exception as e:
                attempt += 1
                error_message = str(e).lower()
                if "api key not valid" in error_message or "connection aborted" in error_message:
                    print(f"\n🔴 Lỗi API/kết nối. Chờ 30s... (lần {attempt}/{max_retries})")
                    time.sleep(30)
                else:
                    print(f"\n🟡 Lỗi xử lý. Thử lại sau 1s... (lần {attempt}/{max_retries})")
                    time.sleep(1)

        print(f"\n❌ Lỗi: Không thể nhận được kết quả hợp lệ sau {max_retries} lần thử.")
        return default_scores

    # --- ĐÃ CẬP NHẬT LẠI CÔNG THỨC TÍNH ĐIỂM TRỌNG SỐ ---
    def process_dataframe(self, input_df: pd.DataFrame) -> pd.DataFrame:
        """Xử lý toàn bộ DataFrame, đánh giá và tính điểm tổng hợp có trọng số."""
        results_list = []

        for index, row in tqdm(input_df.iterrows(), total=input_df.shape[0], desc="Đánh giá các dòng dữ liệu"):
            data_row_dict = row.to_dict()

            scores = self._judge_all_criteria(data_row_dict)

            data_row_dict.update(scores)

            # --- LOGIC TÍNH ĐIỂM CÓ TRỌNG SỐ (8 TIÊU CHÍ) ---
            conciseness = scores.get('tính súc tích', 0)
            clarity = scores.get('tính dễ hiểu', 0)
            coherence = scores.get('tính chặt chẽ', 0)
            relevance = scores.get('tính liên quan', 0)

            # Các tiêu chí có trọng số x2
            correctness = scores.get('tính đúng đắn', 0)
            completeness = scores.get('tính đầy đủ', 0)
            structure = scores.get('Cấu trúc câu trả lời', 0)
            formatting = scores.get('Định dạng câu trả lời', 0) # Tiêu chí mới

            # Công thức tính điểm tổng hợp có trọng số
            # Tổng các trọng số là 1 + 1 + 1 + 1 + 2 + 2 + 2 + 2 = 12
            weighted_score = (conciseness + clarity + coherence + relevance +
                              correctness * 2 + completeness * 2 + structure * 2 + formatting * 2) / 12

            data_row_dict['weighted_score'] = weighted_score

            results_list.append(data_row_dict)

        return pd.DataFrame(results_list)
````

````python
import pandas as pd
from sklearn.metrics import cohen_kappa_score

# Tải 4 file dữ liệu từ Google Colab

liem_qwen = pd.read_csv('/content/Liêm - Qwen__Qwen3-8B.csv')
liem_gemma = pd.read_csv('/content/Liêm - google__gemma-2-2b-it.csv')
dang_qwen = pd.read_csv('/content/Đăng - Qwen__Qwen3-8B.csv')
dang_gemma = pd.read_csv('/content/Đăng - google__gemma-2-2b-it.csv')
print("✅ Đã tải thành công 4 file dữ liệu.")

# Định nghĩa danh sách 8 cột tiêu chí cần tính Kappa
# Đây là nguồn dữ liệu duy nhất để code biết cần phân tích cột nào.
criteria_columns = [
    'tính dễ hiểu',
    'tính đúng đắn',
    'Cấu trúc câu trả lời',
    'Định dạng câu trả lời'
]
````

````python
def calculate_kappa_for_model(rater1_df, rater2_df, columns_to_check):
    """
    Hàm này tính toán chỉ số Cohen's Kappa cho một danh sách các tiêu chí
    giữa hai người đánh giá (đại diện bởi hai DataFrame).

    Args:
        rater1_df (pd.DataFrame): DataFrame chứa điểm của người 1.
        rater2_df (pd.DataFrame): DataFrame chứa điểm của người 2.
        columns_to_check (list): Danh sách tên các cột tiêu chí cần tính.

    Returns:
        dict: Một dictionary với tên tiêu chí là key và điểm Kappa là value.
    """
    kappa_results = {}
    print(f"Bắt đầu tính toán cho {len(columns_to_check)} tiêu chí...")

    for criterion in columns_to_check:
        # Lấy ra cột điểm của người 1 và người 2 cho tiêu chí hiện tại
        ratings1 = rater1_df[criterion]
        ratings2 = rater2_df[criterion]

        # Sử dụng hàm cohen_kappa_score từ thư viện scikit-learn
        # Hàm này sẽ tự động tính toán Po và Pe để cho ra kết quả cuối cùng.
        kappa_score = cohen_kappa_score(ratings1, ratings2)

        # Lưu kết quả vào dictionary
        kappa_results[criterion] = kappa_score

    print("✅ Hoàn thành tính toán.")
    return kappa_results

# --- BẮT ĐẦU THỰC THI ---

# 1. Tính Kappa cho model Qwen giữa Liêm và Đăng
print("--- Model: Qwen ---")
kappa_qwen = calculate_kappa_for_model(liem_qwen, dang_qwen, criteria_columns)

# 2. Tính Kappa cho model Gemma giữa Liêm và Đăng
print("\n--- Model: Gemma ---")
kappa_gemma = calculate_kappa_for_model(liem_gemma, dang_gemma, criteria_columns)


# 3. Trình bày kết quả dưới dạng bảng cho dễ so sánh
results_df = pd.DataFrame({
    "Cohen's Kappa (Qwen)": kappa_qwen,
    "Cohen's Kappa (Gemma)": kappa_gemma
})

print("\n\n--- KẾT QUẢ ĐO LƯỜNG ĐỘ ĐỒNG THUẬN GIỮA HAI NGƯỜI ĐÁNH GIÁ ---")
display(results_df)
````

````python
def calculate_pearson_r_for_model(rater1_df, rater2_df, columns_to_check):
    """
    Hàm này tính toán hệ số tương quan Pearson (r) cho một danh sách các tiêu chí
    giữa hai người đánh giá (đại diện bởi hai DataFrame).

    Args:
        rater1_df (pd.DataFrame): DataFrame chứa điểm của người 1.
        rater2_df (pd.DataFrame): DataFrame chứa điểm của người 2.
        columns_to_check (list): Danh sách tên các cột tiêu chí cần tính.

    Returns:
        dict: Một dictionary với tên tiêu chí là key và điểm Pearson's r là value.
    """
    r_results = {}
    print(f"Bắt đầu tính toán Pearson's r cho {len(columns_to_check)} tiêu chí...")

    for criterion in columns_to_check:
        # Lấy ra cột điểm của người 1 và người 2 cho tiêu chí hiện tại
        ratings1 = rater1_df[criterion]
        ratings2 = rater2_df[criterion]

        # Sử dụng phương thức .corr() sẵn có của pandas để tính Pearson's r
        # Đây là cách tính hiệu quả và chuẩn xác.
        r_score = ratings1.corr(ratings2)

        # Lưu kết quả vào dictionary
        r_results[criterion] = r_score

    print("✅ Hoàn thành tính toán.")
    return r_results

# --- BẮT ĐẦU THỰC THI ---

# 1. Tính Pearson's r cho model Qwen giữa Liêm và Đăng
print("--- Model: Qwen (Pearson's r) ---")
r_qwen = calculate_pearson_r_for_model(liem_qwen, dang_qwen, criteria_columns)

# 2. Tính Pearson's r cho model Gemma giữa Liêm và Đăng
print("\n--- Model: Gemma (Pearson's r) ---")
r_gemma = calculate_pearson_r_for_model(liem_gemma, dang_gemma, criteria_columns)


# 3. Trình bày kết quả dưới dạng bảng cho dễ so sánh
r_results_df = pd.DataFrame({
    "Pearson's r (Qwen)": r_qwen,
    "Pearson's r (Gemma)": r_gemma
})

print("\n\n--- KẾT QUẢ ĐO LƯỜNG TƯƠNG QUAN GIỮA HAI NGƯỜI ĐÁNH GIÁ ---")
display(r_results_df)
````

````python
def find_disagreements(rater1_df, rater2_df, columns_to_check, id_column='id'):
    """
    Hàm này tìm và hiển thị các hàng có sự bất đồng quan điểm giữa hai người đánh giá.

    Args:
        rater1_df (pd.DataFrame): DataFrame điểm của người 1.
        rater2_df (pd.DataFrame): DataFrame điểm của người 2.
        columns_to_check (list): Danh sách các cột tiêu chí để so sánh.
        id_column (str): Cột định danh để ghép hai DataFrame.

    Returns:
        pd.DataFrame: Một DataFrame chỉ chứa các hàng có ít nhất một điểm khác biệt,
                      được định dạng để dễ so sánh.
    """
    # 1. Chọn tất cả các cột context cần thiết
    response_col = 'response' if 'response' in rater1_df.columns else 'reponse'
    context_cols = [id_column, 'question', 'context', 'options', 'ans', 'benchmark', response_col] # <--- THÊM VÀO

    rater1_subset = rater1_df[context_cols + columns_to_check]
    rater2_subset = rater2_df[context_cols + columns_to_check]

    # 2. Gộp hai DataFrame dựa trên cột ID
    merged_df = pd.merge(
        rater1_subset,
        rater2_subset,
        on=id_column,
        how='inner',
        suffixes=('_Liem', '_Dang')
    )

    # 3. Tạo điều kiện lọc để tìm bất đồng
    disagreement_mask = pd.Series([False] * len(merged_df))
    for col in columns_to_check:
        disagreement_mask = disagreement_mask | (merged_df[f'{col}_Liem'] != merged_df[f'{col}_Dang'])

    # 4. Áp dụng điều kiện lọc
    disagreements = merged_df[disagreement_mask].copy()

    # 5. Sắp xếp lại các cột, thêm 'benchmark' vào thứ tự
    final_column_order = [
        id_column,
        'question_Liem',
        'context_Liem',
        'options_Liem',
        'ans_Liem',
        'benchmark_Liem', # <--- THÊM VÀO
        f'{response_col}_Liem'
    ]
    for col in columns_to_check:
        final_column_order.append(f'{col}_Liem')
        final_column_order.append(f'{col}_Dang')

    # Đổi tên các cột context lại cho gọn gàng
    disagreements = disagreements.rename(columns={
        'question_Liem': 'question',
        'context_Liem': 'context',
        'options_Liem': 'options',
        'ans_Liem': 'ans',
        'benchmark_Liem': 'benchmark', # <--- THÊM VÀO
        f'{response_col}_Liem': 'response'
    })

    # Cập nhật lại tên trong danh sách thứ tự
    final_column_order[1] = 'question'
    final_column_order[2] = 'context'
    final_column_order[3] = 'options'
    final_column_order[4] = 'ans'
    final_column_order[5] = 'benchmark' # <--- THÊM VÀO
    final_column_order[6] = 'response'

    # Trả về DataFrame đã được sắp xếp và có đủ các cột mong muốn
    return disagreements[final_column_order]

# --- THỰC THI TÌM KIẾM (giữ nguyên, không cần thay đổi) ---

# Tìm các điểm bất đồng cho model Qwen
print("--- 🕵️ Phân tích bất đồng cho model Qwen ---")
disagreements_qwen = find_disagreements(liem_qwen, dang_qwen, criteria_columns)
print(f"Tìm thấy {len(disagreements_qwen)}/{len(liem_qwen)} trường hợp có bất đồng quan điểm.")
display(disagreements_qwen)

# Tìm các điểm bất đồng cho model Gemma
print("\n--- 🕵️ Phân tích bất đồng cho model Gemma ---")
disagreements_gemma = find_disagreements(liem_gemma, dang_gemma, criteria_columns)
print(f"Tìm thấy {len(disagreements_gemma)}/{len(liem_gemma)} trường hợp có bất đồng quan điểm.")
display(disagreements_gemma)
````

````python
import time
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# --- 1. Khởi tạo Tokenizer và Model ---
# Giả sử bạn đã có tokenizer và model được tải về
# Ví dụ:
# tokenizer = AutoTokenizer.from_pretrained("your-model-name")
# model = AutoModelForCausalLM.from_pretrained("your-model-name")
# if torch.cuda.is_available():
#     model.to("cuda")

# --- 2. Chuẩn bị Input ---
messages = [
    {"role": "user",
     "content": """Bạn là một chuyên gia về lịch sử Việt Nam, cho ngữ cảnh lịch sử, câu hỏi và các lựa chọn, nhiệm vụ của bạn là đưa ra phương án đúng và giải thích tại sao phương án đó đúng dựa trên ngữ cảnh được cung cấp
     ngữ cảnh: Chiến dịch Biên giới thu-đông (1950): Chiến dịch Biên giới được Đảng và Chính phủ quyết định mở vào tháng 6 năm 1950 nhằm tiêu diệt một bộ phận quan trọng sinh lực địch, khai thông đường liên lạc với Trung Quốc và thế giới, đồng thời mở rộng và củng cố căn cứ địa Việt Bắc. Chiến dịch mở màn bằng trận đánh vào vị trí Đông Khê vào sáng 16 tháng 9 năm 1950. Sau khi tiêu diệt Đông Khê, quân địch ở Thất Khê bị uy hiếp và Cao Bằng bị cô lập, buộc Pháp phải rút khỏi Cao Bằng theo Đường số 4. Quân ta đã đoán trước được ý định của địch và chủ động mai phục, chặn đánh địch nhiều nơi trên Đường số 4 khiến cho hai cánh quân này không gặp được nhau. Sự thay đổi chiến thuật từ tấn công trực diện sang mai phục này thể hiện rõ sự linh hoạt và sáng tạo trong tác chiến. Chiến dịch kết thúc thắng lợi, giúp quân đội ta giành được thế chủ động trên chiến trường chính (Bắc Bộ).
    Chiến dịch lịch sử Điện Biên Phủ (1954): Điện Biên Phủ là một tập đoàn cứ điểm mạnh nhất Đông Dương, được các tướng lĩnh Pháp và Mĩ coi là “một pháo đài bất khả xâm phạm”. Mục tiêu của chiến dịch là tiêu diệt lực lượng địch ở đây, giải phóng vùng Tây Bắc, tạo điều kiện giải phóng Bắc Lào. Quân ta nổ súng tấn công tập đoàn cứ điểm Điện Biên Phủ vào ngày 13 tháng 3 năm 1954. Chiến dịch được chia thành ba đợt tấn công liên tiếp. Trong đợt 2 của chiến dịch, khi gặp khó khăn về tiếp tế, ta đã kịp thời khắc phục để nâng cao quyết tâm giành thắng lợi. Điều này cho thấy khả năng điều chỉnh và sáng tạo trong phương án tác chiến để đối phó với tình hình thực tế, đặc biệt khi phải đối mặt với một tập đoàn cứ điểm kiên cố. Ngoài ra, các chiến trường toàn quốc đã phối hợp chặt chẽ nhằm phân tán, tiêu hao, kìm chân địch, tạo điều kiện cho Điện Biên Phủ giành thắng lợi. Chiến dịch kết thúc vào ngày 7 tháng 5 năm 1954 với việc tướng Đờ Caxtơri cùng toàn bộ Ban Tham mưu của địch đầu hàng và bị bắt sống.
     câu hỏi: Chiến dịch Biên giới thu-đông (1950) và chiến dịch Điện Biên Phủ (1954) ở Việt Nam có điểm giống nhau nào sau đây?
     các lựa chọn: ['Có sự linh hoạt, sáng tạo trong phương án tác chiến.', 'Đánh bại cuộc tấn công của Pháp vào cơ quan đầu não kháng chiến.', 'Có ý nghĩa quyết định đến thắng lợi cuối cùng trên mặt trận ngoại giáo.', 'Sử dụng cách đánh công kiên vào tập đoàn cứ điểm.']
     """},
]

# Khi sử dụng device_map, không cần chỉ định .to(device) cho input
inputs = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,
    tokenize=True,
    return_dict=True,
    return_tensors="pt",
)

# Chuyển input lên GPU nếu có
if torch.cuda.is_available():
    inputs = {k: v.to("cuda") for k, v in inputs.items()}

# --- 3. Đo thời gian và chạy Inference ---
start_time = time.time()

outputs = model.generate(**inputs, max_new_tokens=2048)

# Thêm synchronize để đảm bảo tất cả các tác vụ trên GPU đã hoàn thành trước khi dừng đồng hồ
if torch.cuda.is_available():
    torch.cuda.synchronize()

end_time = time.time()

# --- 4. Xử lý và in kết quả ---
inference_time = end_time - start_time
print(f"\nThời gian inference của model: {inference_time:.4f} giây")

# Lấy số token của input
num_input_tokens = inputs["input_ids"].shape[-1]

# Lấy chuỗi token của output (không bao gồm input)
output_tokens = outputs[0][num_input_tokens:]
num_output_tokens = len(output_tokens)

print(f"Số token của output: {num_output_tokens}")

# In ra phản hồi của model
print("\n--- Phản hồi của Model ---")
print(tokenizer.decode(output_tokens))
````
